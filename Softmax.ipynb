Softmax Calculation 

Using an array and a matrix



```python
import numpy as np
```

 Use Randn to return a sample (or samples) from the “standard normal” distribution.

```python
a = np.random.randn(5)
a
```

Step 1: We want to exponentiate these. Remember we exponentiate to turn the elements positive

```python
expa = np.exp(a)
expa
```

Step 2: We want the answer to be each element divided by sum of elements

```python
answer = expa / expa.sum()
answer
```

The sum of answers should equal to 1

```python
answer.sum()
```

Lets try this again with a matrix

```python
A = np.random.rand(100,5) #This is 100 samples and 5 classes
A
```

```python
expA = np.exp(A)
expA
```

Now we have a problem if we do the same thing as last time because the whole matrix will sum to 1.
We want every row to sum to 1

```python
#Solution

answerA = expA / expA.sum(axis = 1, keepdims=True)
```

Note for keepdims: If this is set to True, the axes which are reduced are left in the result as dimensions with size one. 
    With this option, the result will broadcast correctly against the input array.

```python
answerA
```

```python
# To check

answerA.sum(axis=1)
```

```python
# Check to see what denominator gives you

expA.sum(axis=1, keepdims=True)
```
